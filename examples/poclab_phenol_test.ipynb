```python
import pandas as pd
from alfabet import model
```

```python
test_data = pd.read_csv('試薬リスト.csv')
```

```python
test_data.detail()
```

```python
test_data.head()
```

```python
#smilesの項目のuniqueなものだけを抽出する。
poclabphenollist = test_data.smiles.unique()
poclabphenollist[:5]
```

```python
df = model.predict(poclabphenollist, verbose=False)
```

```python
#項目名
df.columns.values
```

```python
df
```

```python
# お試しで、BDEの予測値が85以下で抽出
df_lower85 = df[df['bde_pred'] < 85]
df_lower85.head()
```

```python
# 
df_lower85.sort_values(["bde_pred"]).head()
```

```python
# BDEの予測値をH-Oだけで抽出
df_OH = df[df['bond_type'] == 'H-O']
df_OH.head()
```

```python
df.columns
```

```python
# 必要なものだけにする。
df_OH_limit = df_OH[['molecule', 'bond_index', 'fragment2','bde_pred',]]
df_OH_limit.head()
```

```python
from rdkit import Chem
```

```python
df_OH_limit['mol'] = df_OH_limit.molecule.map(Chem.MolFromSmiles)
```
